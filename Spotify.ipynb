1. what is distribution of streams across the dataset?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df.loc[:, 'streams'] = pd.to_numeric(df['streams'], errors='coerce')
sns.histplot(df['streams'].dropna(), kde=True)
plt.xlabel('Streams')
plt.ylabel('Frequency')
plt.show()
```

# Conclusion
 The distribution might show a right skew with most tracks having lower streams and a few having very high streams.

2. Who are the top 10 artists by total streams?

```python
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

top_10_artists = df.dropna(subset=['streams']).groupby('artist(s)_name')['streams'].sum().nlargest(10)

top_10_artists.plot(kind='barh', color='skyblue', figsize=(8, 5))
plt.title('Top 10 Artists by Total Streams', fontsize=14)
plt.xlabel('Total Streams', fontsize=12)
plt.ylabel('Artist', fontsize=12)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

# Conclusion
The bar plot reveals which artists dominate the streaming scene, indicating their popularity and market presence.

3. Do certain months yield more popular songs?

```python
streams_by_month = df.groupby('released_month')['streams'].sum()

plt.figure(figsize=(10, 6))
streams_by_month.plot(kind='bar')
plt.title('Total Streams by Released Month')
plt.xlabel('Month')
plt.ylabel('Total Streams')
plt.xticks(rotation=45)
plt.show()

print(streams_by_month)
```

# Conclusion
If certain months show significantly higher streams, it may suggest seasonal trends in music releases

4. Average Streams by Released Year.

```python
average_streams_by_year = df.groupby('released_year')['streams'].mean()

plt.figure(figsize=(10, 6))
average_streams_by_year.plot(kind='line', marker='o')
plt.title('Average Streams per Song by Released Year')
plt.xlabel('Year')
plt.ylabel('Average Streams')
plt.xticks(rotation=45)
plt.grid()
plt.show()

print(average_streams_by_year)
```

# Conclusion
An upward trend in average streams indicates growing popularity of streaming over the years.

5. What is the distribution of danceability across all songs?

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['danceability_%'], bins=30, kde=True)
plt.title('Distribution of Danceability')
plt.xlabel('Danceability (%)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

# Conclusion
 The histogram indicates common danceability levels, reflecting listener preferences for certain types of songs.

6.  How does instrumentalness affect streaming numbers?

```python
instrumentalness_correlation = df['instrumentalness_%'].corr(df['streams'])

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='instrumentalness_%', y='streams')
plt.title('Instrumentalness vs. Total Streams')
plt.xlabel('Instrumentalness (%)')
plt.ylabel('Total Streams')
plt.axhline(y=df['streams'].mean(), color='r', linestyle='--', label='Average Streams')
plt.legend()
plt.show()

print(f'Instrumentalness Correlation: {instrumentalness_correlation:.2f}')
```

# Conclusion
 A low correlation indicates that instrumental tracks may not perform as well as vocal tracks in terms of streams.

7. What are the most common musical keys in the dataset?

```python
key_counts = df['key'].value_counts()

plt.figure(figsize=(10, 6))
key_counts.plot(kind='bar')
plt.title('Most Common Keys in Songs')
plt.xlabel('Key')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()

print(key_counts)
```

# Conclusion
 The key distribution reveals common musical trends among artists.

8. How do songs in different modes (major/minor) perform in terms of streams?

```python
streams_by_mode = df.groupby('mode')['streams'].mean()

plt.figure(figsize=(10, 6))
streams_by_mode.plot(kind='bar')
plt.title('Average Streams by Musical Mode')
plt.xlabel('Mode (Major/Minor)')
plt.ylabel('Average Streams')
plt.xticks(rotation=45)
plt.show()

print(streams_by_mode)
```

# Conclusion
 Differences in average streams by mode suggest listener preferences for major or minor keys.

9. How does streaming success vary by genre and release year?

```python
streams_by_year_artist = df.groupby(['released_year', 'artist_count'])['streams'].sum().unstack()

plt.figure(figsize=(12, 6))
streams_by_year_artist.plot(kind='bar', stacked=True, colormap='tab20', ax=plt.gca())
plt.title('Total Streams by Release Year and Artist Count')
plt.xlabel('Release Year')
plt.ylabel('Total Streams')
plt.legend(title='Number of Artists', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Conclusion
This stacked bar chart will reveal how different genres have performed over the years in terms of streaming. Notable trends can indicate shifts in listener preferences or the rise of certain genres.

10. What is the average number of streams per song by release year?

```python
average_streams_by_year = df.groupby('released_year')['streams'].mean()

# Plotting
plt.figure(figsize=(10, 6))
average_streams_by_year.plot(kind='bar', color='skyblue')
plt.title('Average Streams per Song by Released Year')
plt.xlabel('Release Year')
plt.ylabel('Average Streams')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

print(average_streams_by_year)
```

# Conclusion
This bar chart displays the average number of streams per song for each release year. An upward trend indicates increasing popularity of songs over the years, reflecting changes in listener habits or the rise of streaming platforms.
